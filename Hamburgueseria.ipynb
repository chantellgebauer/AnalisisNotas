```python
inventario = 100
print(f"El inventario contiene {inventario} hamburguesas")
while inventario > 0:
  if inventario <= 10:
    print("ADVERTENCIA: EL INVENTARIO ESTA CASI VACIO")
  num_hamburguesas = int(input("¿Cuantas hamburguesas quiere desea comprar?"))
  if num_hamburguesas > inventario:
    print("No hay suficientes hamburguesas, solo quedan", inventario, "disponibles")
  else:
    inventario -= num_hamburguesas
    print(f"Has comprado", num_hamburguesas, "hamburguesas, Nos quedan", inventario,"hamburguesas")


  print("CODIGO MODIFICADO")
```

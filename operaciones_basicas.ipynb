```python
import pandas as pd
df = pd.read_csv('ventas.csv')
df.head()
```

```python
df['total'] = df["precio"] * df["unidades"]
df.head()
```

```python
df.groupby("producto")["total"].sum()
```

```python
df[df["unidades"] > 3]
```

```python
df["fecha"] = pd.to_datetime(df["fecha"])
df["mes"] = df["fecha"].dt.month
df.head()
```

```python
df.groupby("mes")["total"].sum()
```
